# Creating one hot matrix for the serrounding seq per CpG

Transferring the Probe To Seq DF to a oneHotEncoding representation

```python
# understanding the oneHotEncoding code
seqList = [0,1,2,1,3,4,0,1]
zeros = np.zeros((len(seqList),Conf.numBases),  dtype=int)
zeros[np.arange(len(seqList)),seqList] = int(1)
zeros
```

```python
import csv

def seq_to_one_hot_mtrx(seq, probe):
    if 'N' in seq:
        print(f"N in seq {probe}. replaced with 5th value - 4")
    seq = seq.replace('A','0').replace('a','0').replace('C', '1').replace('c','1').replace('G', '2').replace('g','2').replace('T', '3').replace('t','3').replace('N','4')
    seqList = list(seq)
    seqList = [int(i) for i in seqList]
    one_hot = np.zeros((len(seq),Conf.numBases))
    one_hot[np.arange(len(seq)),seqList] = 1
    return one_hot.flatten()

def sequences_to_one_hot_mtrx(probe_to_serrounding_seq, path_one_hot):
  probe = probe_to_serrounding_seq["Probe"].iloc[0]
  seq = probe_to_serrounding_seq["Seq"].iloc[0]
  oneHotMtrx = seq_to_one_hot_mtrx(seq, str(probe))
  len_seq_one_hot = len(oneHotMtrx.flatten())
  lst = range(len_seq_one_hot)
  probe_to_seq_one_hot = {}
  for i in range(probe_to_serrounding_seq.shape[0]):
      probe = probe_to_serrounding_seq["Probe"].iloc[i]
      seq = probe_to_serrounding_seq["Seq"].iloc[i]
      oneHotMtrx = seq_to_one_hot_mtrx(seq, str(probe))
      probe_to_seq_one_hot[probe] = oneHotMtrx.flatten()
      if i%1000 ==0:
        print(i)
  with open(path_one_hot, "w") as outfile:
      writer = csv.writer(outfile)
      cpg_list = list(probe_to_seq_one_hot.keys())
      lst = range(len_seq_one_hot)
      columns= ['Probe']+[format(x, '02d') for x in lst]
      writer.writerow(columns)
      for i, cpg in enumerate(cpg_list):
          row = [cpg] + list(probe_to_seq_one_hot[cpg])
          writer.writerow(row)

  return probe_to_seq_one_hot
```
